```python
import dataTransformations
import plots
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.tsa.stattools as stats
import statsmodels.graphics.tsaplots as stats_graph
```

```python
df=pd.read_csv("Data1997-2022.csv")
```

```python
fig,axs=plt.subplots(2,2,sharex=True)
axs[0,0].plot(df.iloc[:,1])
axs[0,0].set_title("CPI")
axs[0,0].set_ylabel("CPI index")
axs[0,1].plot(df.iloc[:,2])
axs[0,1].set_ylabel("Kr")
axs[0,1].set_title("Monthly salary")
axs[1,0].plot(df.iloc[:,3])
axs[1,0].set_ylabel("Percentage")
axs[1,0].set_title("Policy rate")
axs[1,0].set_xlabel("Months")
axs[1,1].plot(df.iloc[:,4])
axs[1,1].set_title("Unemployment percentage")
axs[1,1].set_ylabel("Percentage")
axs[1,1].set_xlabel("Months")
plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
acf_CPI=plots.acf(np.array(df.iloc[:,1]),48)

plots.acf_plot(acf_CPI, n = len(np.array(df.iloc[:,1])))
```

```python
acf_CPI=plots.pacf(np.array(df.iloc[:,1]),24)
plots.pacf_plot(acf_CPI, n= np.array(df.iloc[:,1]))
```

```python
cpi_diff=dataTransformations.differencing(df.iloc[:,1],1)
plt.figure()
plt.title("CPI diff 1")
plt.plot(cpi_diff)
plt.show()
```

```python
cpi_diff=dataTransformations.differencing(df.iloc[:,1],12)
plt.figure()
plt.title("CPI difference 12")
plt.plot(cpi_diff)
plt.xlabel("Months")
plt.show()
```

```python
acf_CPI=plots.acf(cpi_diff,48)

plots.acf_plot(acf_CPI, n = len(cpi_diff))
```

```python
pacf_CPI=plots.pacf(cpi_diff,48)


plots.pacf_plot(pacf_CPI, n = len(cpi_diff))
```

```python
epsilon=dataTransformations.differencing(cpi_diff,1)
epsilon=epsilon**2

plt.plot(np.sqrt(epsilon))
```

```python
acf_epsilon=plots.acf(epsilon,48)

plots.acf_plot(acf_epsilon, n = len(epsilon))
```

```python
acf_epsilon=plots.pacf(epsilon,48)

plots.pacf_plot(acf_epsilon, n = len(epsilon))
```
